## Q.1 Can you explain the logic and working of the Tower of Hanoi algorithm by writing a Java program? How does the recursion work, and how are the movements of disks between rods accomplished?


The problem involves three rods and a number of disks of different sizes which can be moved between the rods. The puzzle starts with the disks in a neat stack in ascending order of size on one rod, the smallest at the top.

The rules for Tower of Hanoi are as follows:

- Only one disk can be moved at a time.
- Each move consists of taking the upper disk from one of the stacks and placing it on top of another stack or on an empty rod.
- No disk may be placed on top of a smaller disk.

```python
# Python Program
i = 0
def tower_of_hanoi(n,source, target,auxiliary):
    if n==1:
        print(f"Move disk 1 from {source} to {target}")
    else:
        tower_of_hanoi(n-1, source,auxiliary, target)
        print(f"Move disk {n} from {source} to {target}")
        tower_of_hanoi(n-1, auxiliary, target, source)
        
        
# Driver code 
n= 3
tower_of_hanoi(3,'A', 'C', 'B')
```

- `tower_of_hanoi` is a recursive function that takes the number of disks (`n`) and the names of the source, target, and auxiliary rods.
- The base case is when there is only one disk (`n == 1`), in which case it prints the move from the source rod to the target rod.
- Otherwise, it recursively moves the top `n-1` disks from the source rod to the auxiliary rod, then moves the nth disk from the source rod to the target rod, and finally recursively moves the `n-1` disks from the auxiliary rod to the target rod.


# Q.2 Given two strings word1 and word2, return the minimum number of operations required to convert word1 to word2.
- `Example 1`:

`Input`: word1 = "horse", word2 = "ros"

`Output`: 3

`Explanation`:

horse -> rorse (replace 'h' with 'r')

rorse -> rose (remove 'r')

rose -> ros (remove 'e')

- `Example 2`:

`Input`: word1 = "intention", word2 = "execution"

`Output`: 5

`Explanation`:

intention -> inention (remove 't')

inention -> enention (replace 'i' with 'e')

enention -> exention (replace 'n' with 'x')

exention -> exection (replace 'n' with 'c')

exection -> execution (insert 'u')

```python
def min_operations(word1, word2):
    memo = {}   # Memoization

    def dp(i, j):
        if (i, j) in memo:
            return memo[(i, j)]

        # Base cases: empty strings
        if i == 0:
            return j
        if j == 0:
            return i

        # Recursive cases
        if word1[i - 1] == word2[j - 1]:
            memo[(i, j)] = dp(i - 1, j - 1)
        else:
            memo[(i, j)] = 1 + min(
                dp(i - 1, j),       # Deletion
                dp(i, j - 1),       # Insertion
                dp(i - 1, j - 1)    # Replacement
            )

        return memo[(i, j)]

    return dp(len(word1), len(word2))

# Example usage:
word1_1, word2_1 = "horse", "ros"
word1_2, word2_2 = "intention", "execution"

output_1 = min_operations(word1_1, word2_1)
output_2 = min_operations(word1_2, word2_2)

print("Output 1:", output_1)  # Output: 3
print("Output 2:", output_2)  # Output: 5
```

Q. 3 Print the max value of the array [ 13, 1, -3, 22, 5].

```python
def find_max(arr, n):
    # Base case: if there is only one element, return it
    if n == 1:
        return arr[0]
    
    # Recursive case: find the maximum in the rest of the array
    max_in_rest = find_max(arr[1:], n-1)
    
    # Compare the maximum in the rest with the first element and return the larger of the two
    return max(arr[0], max_in_rest)

# Example array
array = [13, 1, -3, 22, 5]

# Find the maximum using the recursive function
max_value = find_max(array, len(array))

# Print the result
print("The maximum value in the array is:", max_value)
```

Q.4 Find the sum of the values of the array [92, 23, 15, -20, 10].

```python
def sum_values(arr, n):
    # Base case: if there is only one element, return it
    if n == 1:
        return arr[0]
    
    # Recursive case: find the maximum in the rest of the array
    sum_in_rest = sum_values(arr[1:], n-1)
    
    # Compare the maximum in the rest with the first element and return the larger of the two
    return arr[0]+ sum_in_rest 
# Example array
array = [92, 23, 15, -20, 10]

# Find the maximum using the recursive function
sum_values = sum_values(array, len(array))

# Print the result
print("The sum of  value in the array is:", sum_values)
```

Q.5 Given a number n. Print if it is an armstrong number or not.An armstrong number is a number if the sum
of every digit in that number raised to the power of total digits in that number is equal to the number.

Example : 153 = 1^3 + 5^3 + 3^3 = 1 + 125 + 27 = 153 hence 153 is an armstrong number. (Easy)

Input1 : 153

Output1 : Yes

Input 2 : 134

Output2 : No

```python
def check_armstrong(n):
    s = len(str(n))
    if n//10==0:
        return n ==n**s
    else:
        return n== (n%10)**s + check_armstrong(n//10)
input_num = 153
output_result = "Yes" if check_armstrong(input_num) else "No"
print(f"Input: {input_num}\nOutput: {output_result}")
```

```python
def count_digits(n):
    if n == 0:
        return 0
    return 1 + count_digits(n // 10)

def is_armstrong_recursive(n, total_digits=None):
    if total_digits is None:
        total_digits = count_digits(n)

    if n == 0:
        return 0
    else:
        return (n % 10) ** total_digits + is_armstrong_recursive(n // 10, total_digits)

def is_armstrong_number(n):
    total_digits = count_digits(n)
    return n == is_armstrong_recursive(n, total_digits)

# Example usage:
input_number = int(input("Enter a number: "))
if is_armstrong_number(input_number):
    print("Yes")
else:
    print("No")
```
